```python
#pip install pandas-datareader
#pip install altair

import numpy as np
import pandas as pd
from pandas_datareader import data
import altair as alt
```

```python
start = '2020-1-1'
end = '2021-1-5'
source = 'yahoo'
```

```python
apple = data.DataReader("AAPL", start=start, end=end, data_source=source).reset_index()
```

```python
IBM = data.DataReader("IBM", start=start, end=end, data_source=source).reset_index()
```

```python
microsoft = data.DataReader("MSFT", start=start, end=end, data_source=source).reset_index()
```

```python
apple['Symbol']='APPL'
IBM['Symbol'] = 'IBM'
microsoft['Symbol'] = 'MSFT'
```

```python
stocks = pd.concat([apple[['Date','Close','Volume','Symbol']],IBM[['Date','Close','Volume','Symbol']],microsoft[['Date','Close','Volume','Symbol']]],axis=0)
```

```python
(alt.Chart(stocks[stocks.Symbol == 'APPL']).mark_line().encode(x='Date', y='Close'))
```

```python
(alt.Chart(stocks).mark_line().encode(x='Date', y='Close',color='Symbol').properties(height=300, width=500))
```

```python
MSFT_resampled = stocks[stocks.Symbol == 'MSFT'].resample('7D',on='Date').mean().reset_index()
```

```python
(alt.Chart(MSFT_resampled).mark_line().encode(x='Date', y='Close').properties(height=300, width=500))
```

```python
price = (alt.Chart(MSFT_resampled).mark_line().encode(x='Date', y='Close'))
volume = (alt.Chart(stocks[stocks.Symbol =='MSFT']).mark_line().encode(x='Date', y='Volume'))
          
price | volume
```

```python
line = (alt.Chart(stocks).mark_line().encode(x='Date', y='Close',color='Symbol').properties(height=400,width=500))
avg = (alt.Chart(stocks).mark_rule().encode(y='average(Close)',color='Symbol', size=alt.value(2)))         
line + avg
```
